##Exploring the data through SQL queries
note: dates are the closest things to primary keys, so we will be using them as such

```sql
SELECT date, CAST(open AS DECIMAL(10,3)), CAST(high - low AS DECIMAL(10,2)) AS price_change FROM amazon_cleaned WHERE high - low > 0.99 ORDER BY (high-low) DESC
```

Using the above information we see the largest dollar price change in amazon's stock happened July 7th 2020. It is important to identify the overall price when noticing this.

# Compare Amazon and Tesla's opening prices

```sql
SELECT amazon_cleaned.date, CAST(amazon_cleaned.open AS DECIMAL(10,2)) AS Amazon_Open, CAST(tesla_cleaned.open AS DECIMAL(10,2)) AS Tesla_Open 
FROM amazon_cleaned INNER JOIN tesla_cleaned ON amazon_cleaned.date = tesla_cleaned.date ORDER BY amazon_cleaned.date ASC 
```

# Amazon's average price change grouped by month

```sql
SELECT 
  MONTH(date) AS Month,
  AVG(CAST(((close - open)/open * 100) AS DECIMAL(10,3))) AS Average_Percent_Change
FROM amazon_cleaned
GROUP BY Month
ORDER BY Average_Percent_Change DESC;
```

This data is fairly interesting given that the January effect is a theorized anomoly viewed in the stock market. While our data does not show this it does show the largest decrease in December, meaning that a increase in January is evident.

```sql
SELECT 
  MONTH(date) AS Month,
  EXP(AVG(LOG(1 + (close - open)/open))) AS geometric_close
FROM amazon_cleaned
GROUP BY Month
ORDER BY geometric_close ASC;
```

Above we used the geometric calculation to see if the data still shows a trend when compounding of percent changes are taken into thought.

#More Explorations but With All The Stocks

First we will find average monthly returns for all the stock tables (keep in mind the data may be slightly incorrect since some stocks have been around longer).

```sql
SELECT 
  MONTH(amazon_cleaned.date) AS Month,
  AVG(CAST(((amazon_cleaned.close - amazon_cleaned.open) / amazon_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  AMZN_pc,
  AVG(CAST(((apple_cleaned.close - apple_cleaned.open) / apple_cleaned.open * 100) AS DECIMAL(10))) AS 
  APPL_pc,
  AVG(CAST(((berkshire_b_cleaned.close - berkshire_b_cleaned.open) / berkshire_b_cleaned.open * 100) AS DECIMAL(10))) AS BRKB_pc,
  AVG(CAST(((eli_lily_cleaned.close - eli_lily_cleaned.open) / eli_lily_cleaned.open * 100) AS DECIMAL(10))) AS 
  LLY_pc,
  AVG(CAST(((jp_morgan_cleaned.close - jp_morgan_cleaned.open) / jp_morgan_cleaned.open * 100) AS DECIMAL(10))) AS 
  JPM_pc,
  AVG(CAST(((microsoft_cleaned.close - microsoft_cleaned.open) / microsoft_cleaned.open * 100) AS DECIMAL(10))) AS 
  MSFT_pc,
  AVG(CAST(((tesla_cleaned.close - tesla_cleaned.open) / tesla_cleaned.open * 100) AS DECIMAL(10))) AS 
  TSLA_pc,
  AVG(CAST(((united_health_cleaned.close - united_health_cleaned.open) / united_health_cleaned.open * 100) AS DECIMAL(10))) AS UNH_pc
FROM
  amazon_cleaned INNER JOIN apple_cleaned ON amazon_cleaned.date = apple_cleaned.date INNER JOIN berkshire_b_cleaned ON amazon_cleaned.date = berkshire_b_cleaned.date INNER JOIN eli_lily_cleaned ON amazon_cleaned.date = eli_lily_cleaned.date INNER JOIN jp_morgan_cleaned ON amazon_cleaned.date = jp_morgan_cleaned.date INNER JOIN microsoft_cleaned ON amazon_cleaned.date = microsoft_cleaned.date INNER JOIN tesla_cleaned ON amazon_cleaned.date = tesla_cleaned.date INNER JOIN united_health_cleaned ON amazon_cleaned.date = united_health_cleaned.date
GROUP BY Month
ORDER BY Month ASC;
```

Our data is slightly different because the dates had to allign for this query to work. However, it still yields some interesting results. 7/8 Stocks have average negative returns in december and all stocks have positive returns in January.

##Similar Query But grouping by year instead of month
(since our data only goes up to the beginning of 2024 it should not be analyzed for 2024)

```sql
SELECT 
  YEAR(amazon_cleaned.date) AS YEAR,
  SUM(CAST(((amazon_cleaned.close - amazon_cleaned.open) / amazon_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  AMZN_pc,
  SUM(CAST(((apple_cleaned.close - apple_cleaned.open) / apple_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  APPL_pc,
  SUM(CAST(((berkshire_b_cleaned.close - berkshire_b_cleaned.open) / berkshire_b_cleaned.open * 100) AS DECIMAL(10,2))) AS BRKB_pc,
  SUM(CAST(((eli_lily_cleaned.close - eli_lily_cleaned.open) / eli_lily_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  LLY_pc,
  SUM(CAST(((jp_morgan_cleaned.close - jp_morgan_cleaned.open) / jp_morgan_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  JPM_pc,
  SUM(CAST(((microsoft_cleaned.close - microsoft_cleaned.open) / microsoft_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  MSFT_pc,
  SUM(CAST(((tesla_cleaned.close - tesla_cleaned.open) / tesla_cleaned.open * 100) AS DECIMAL(10,2))) AS 
  TSLA_pc,
  SUM(CAST(((united_health_cleaned.close - united_health_cleaned.open) / united_health_cleaned.open * 100) AS DECIMAL(10,2))) AS UNH_pc
FROM
  amazon_cleaned LEFT JOIN apple_cleaned ON amazon_cleaned.date = apple_cleaned.date LEFT JOIN berkshire_b_cleaned ON amazon_cleaned.date = berkshire_b_cleaned.date LEFT JOIN eli_lily_cleaned ON amazon_cleaned.date = eli_lily_cleaned.date LEFT JOIN jp_morgan_cleaned ON amazon_cleaned.date = jp_morgan_cleaned.date LEFT JOIN microsoft_cleaned ON amazon_cleaned.date = microsoft_cleaned.date LEFT JOIN tesla_cleaned ON amazon_cleaned.date = tesla_cleaned.date LEFT JOIN united_health_cleaned ON amazon_cleaned.date = united_health_cleaned.date
GROUP BY Year
ORDER BY Year ASC;
```

If you believe the past is a somewhat realistic view of the future (so stocks are not a random walk). Then you can use this information to try and time markets. If you see the highest percentage change in a year for a stock is around 40% and the stock is already up 40% you might agree to not invest at the moment.

###Calculating the total average from all the years to get an all year average 

```sql
%python
# Run the SQL query and load the results into a Pandas DataFrame to plot with mathplotlib
query = """
SELECT 
  YEAR(amazon_cleaned.date) AS Year,
  SUM(((amazon_cleaned.close - amazon_cleaned.open) / amazon_cleaned.open * 100)) AS AMZN_pc,
  SUM(((apple_cleaned.close - apple_cleaned.open) / apple_cleaned.open * 100)) AS APPL_pc,
  SUM(((berkshire_b_cleaned.close - berkshire_b_cleaned.open) / berkshire_b_cleaned.open * 100)) AS BRKB_pc,
  SUM(((eli_lily_cleaned.close - eli_lily_cleaned.open) / eli_lily_cleaned.open * 100)) AS LLY_pc,
  SUM(((jp_morgan_cleaned.close - jp_morgan_cleaned.open) / jp_morgan_cleaned.open * 100)) AS JPM_pc,
  SUM(((microsoft_cleaned.close - microsoft_cleaned.open) / microsoft_cleaned.open * 100)) AS MSFT_pc,
  SUM(((tesla_cleaned.close - tesla_cleaned.open) / tesla_cleaned.open * 100)) AS TSLA_pc,
  SUM(((united_health_cleaned.close - united_health_cleaned.open) / united_health_cleaned.open * 100)) AS UNH_pc
FROM
  amazon_cleaned 
LEFT JOIN 
  apple_cleaned ON amazon_cleaned.date = apple_cleaned.date 
LEFT JOIN 
  berkshire_b_cleaned ON amazon_cleaned.date = berkshire_b_cleaned.date 
LEFT JOIN 
  eli_lily_cleaned ON amazon_cleaned.date = eli_lily_cleaned.date 
LEFT JOIN 
  jp_morgan_cleaned ON amazon_cleaned.date = jp_morgan_cleaned.date 
LEFT JOIN 
  microsoft_cleaned ON amazon_cleaned.date = microsoft_cleaned.date 
LEFT JOIN 
  tesla_cleaned ON amazon_cleaned.date = tesla_cleaned.date 
LEFT JOIN 
  united_health_cleaned ON amazon_cleaned.date = united_health_cleaned.date
GROUP BY 
  Year
ORDER BY 
  Year ASC
"""


all_yearly_df = spark.sql(query).toPandas()
```

```sql
%python
import matplotlib.pyplot as plt
import numpy as np
from pyspark.sql.functions import avg, col

y = []
y.append(all_yearly_df['AMZN_pc'].mean())
y.append(all_yearly_df['APPL_pc'].mean())
y.append(all_yearly_df['BRKB_pc'].mean())
y.append(all_yearly_df['LLY_pc'].mean())
y.append(all_yearly_df['JPM_pc'].mean())
y.append(all_yearly_df['MSFT_pc'].mean())
y.append(all_yearly_df['TSLA_pc'].mean())
y.append(all_yearly_df['UNH_pc'].mean())

x = np.arange(len(y))
labels = ['Amazon', 'Apple', 'Berkshire B', 'Eli Lilly', 'JP Morgan', 'Microsoft', 'Tesla', 'United Health']

fig, ax = plt.subplots()
ax.bar(x, y, width=1, edgecolor='white',linewidth = 0.7)

ax.set_xlim(-0.5, len(y) -0.5)
ax.set_ylim(min(y) - 10, max(y) + 10)
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45, ha='center')
ax.set_yticks(np.arange(int(min(y) - 10), int(max(y) + 10), 10))

ax.set_title('Average Total Yearly Percent Change')
ax.set_ylabel('Average Total Yearly Deviation in Percentage')

plt.show()
```

This might not be the most reliable information since it is an arithmetic average and we are talking about percentages, however it is interesting information. This is also data day by day grouped by month, which is much different than looking at the stock price a year or 5 ago.
